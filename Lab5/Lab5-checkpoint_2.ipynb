```python
import pandas as pd
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
from pyedflib import highlevel
```

```python
signals_1, signal_headers_1, header_1 = highlevel.read_edf('Subject00_1.edf')
signals_2, signal_headers_2, header_2 = highlevel.read_edf('Subject00_2.edf')
```

```python
s1_o1 = np.fft.fft(signals_1[14])
s1_o2 = np.fft.fft(signals_1[15])
s2_o1 = np.fft.fft(signals_2[14])
s2_o2 = np.fft.fft(signals_2[15])
#O1 - 14, o2 - 15  
```

```python
#500 отметок в секунду
i=0
for s in s1_o1:
    i +=1
print('Датчик О1 запись 1 имеет ',i, ' данных')
i=0
for s in s1_o2:
    i +=1
print('Датчик О2 запись 1 имеет ',i, ' данных')
i=0
for s in s2_o1:
    i +=1
print('Датчик О1 запись 2 имеет ',i, ' данных')
i=0
for s in s2_o2:
    i +=1
print('Датчик О2 запись 2 имеет ',i, ' данных')
```

```python
#ХРЕНЬ
n_s1_o1 = np.arange(181, dtype=float)
n_s1_o2 = np.arange(181, dtype=float)
n_s2_o1 = np.arange(61, dtype=float)
n_s2_o2 = np.arange(61, dtype=float)
k=0
tmp = 0
s1=0.0
s2=0.0
for i in range(91000):
    tmp += 1
    if (tmp<=500):
        s1 +=s1_o1[i]
        s2 +=s1_o2[i]
    else:
        n_s1_o1[k]=abs(s1/500)
        n_s1_o2[k]=abs(s2/500)
        k += 1
        tmp = 0
s1=0.0
s2=0.0
k=0
tmp=0
for i in range(31000):
    tmp += 1
    if (tmp<=500):
        s1 +=s2_o1[i]
        s2 +=s2_o2[i]
    else:
        n_s2_o1[k]=abs(s1/500)
        n_s2_o2[k]=abs(s2/500)
        k += 1
        tmp = 0
```
